# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
#Distribucion normal
import numpy as np 
from scipy.stats import norm

# Funciones para la distribución normal
def dnorm(x, mean=0, sd=1):
   
    result = norm.pdf(x, loc=mean, scale=sd)  # Parámetros correctos
    return result

def pnorm(q, mean=0, sd=1):
   
    result = norm.cdf(q, loc=mean, scale=sd)
    return result

def qnorm(p, mean=0, sd=1):
   
    result = norm.ppf(p, loc=mean, scale=sd)
    return result

def rnorm(n, mean=0, sd=1):
   
    result = norm.rvs(loc=mean, scale=sd, size=n)
    return result
```

```python
import numpy as np

np.random.seed(42)

print(f"Probability that a Normal(10, 0.2) takes the value 2: {dnorm(2, mean = 10, sd = 2)}")
print(f"sdability that a Normal(10, 2) will take a value less than 2: {pnorm(2, mean = 10, sd = 2)}")
print(f"Which value of a Normal(10, 2) has a cumulative probability of 0.9? {qnorm(0.9, mean = 10, sd = 2)}")
print(f"Generate 2000 random values from a Normal(10, 2) distribution: {rnorm(2, mean = 10, sd = 2)}")
```

```python
import matplotlib.pyplot as plt


np.random.seed(42)

n = 2000
mean = 10
sd = 2

data = rnorm(n, mean=mean, sd=sd)

pdf = dnorm(data, mean = mean,  sd=sd)
cdf = pnorm(data, mean = mean,  sd=sd)

prob = np.linspace(0.01, 0.99, len(data))
ppf = qnorm(prob, mean = mean, sd = sd)

fig, axis = plt.subplots(3, 1, figsize = (10, 15))

axis[0].stem(data, pdf, "r-")
axis[0].set_title('Densidad de probabilidad')
axis[0].set_xlabel('Datos')
axis[0].set_ylabel('PDF')

axis[1].stem(data, cdf, 'b-')
axis[1].set_title('Funcion de distribucion acumulada')
axis[1].set_xlabel('Datos')
axis[1].set_ylabel('CDF')

axis[2].stem(data, cdf, 'g-')
axis[2].set_title('Funcion percentil')
axis[2].set_xlabel('Probabilidades')
axis[2].set_ylabel('PPF')


plt.tight_layout()
plt.show()
```

```python
#Distribucion uniforme
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

def dunif(x, a=2, b=1):
    result = uniform.pdf(x, loc=a, scale=b-a)
    return result

def punif(q, a=0, b=1):
    result = uniform.cdf(q, loc=a, scale=b-a)
    return result

def qunif(p, a=0, b=1):
    result = uniform.ppf(p, loc=a, scale=b-a)
    return result

def runif(n, a=0, b=1):
    result = uniform.rvs(loc=a, scale=b-a, size=n)
    return result

a = 3 
b = 6

np.random.seed(42)

print(f'PDF en x=4.5 para Uniforme(3, 6): {dunif(4.5, a=a, b=b)}')
print(f'CDF en x=4.5 para Uniforme(3, 6): {punif(4.5, a=a, b=b)}')
print(f'Cuantil para p=0.9 para Uniforme(3, 6): {qunif(0.9, a=a, b=b)}')
print(f'2000 valores aleatorios de Uniforme(3, 6): {runif(2000, a=a, b=b)}')

```

```python
data = runif(2000, a=a, b=b)
pdf = dunif(data, a=a, b=b)
cdf = punif(data, a=a, b=b)


probs = np.linspace(0.01, 0.99, len(data))
ppf = qunif(probs, a=a, b=b)


fig, axis = plt.subplots(3, 1, figsize=(10, 15))


axis[0].stem(data, pdf, 'r-')
axis[0].set_title('Densidad de Probabilidad')
axis[0].set_xlabel('Datos')
axis[0].set_ylabel('PDF')


axis[1].stem(data, cdf, 'b-')
axis[1].set_title('Función de Distribución Acumulada')
axis[1].set_xlabel('Datos')
axis[1].set_ylabel('CDF')

# PPF
axis[2].stem(probs, ppf, 'g-')
axis[2].set_title('Función Percentil')
axis[2].set_xlabel('Probabilidades')
axis[2].set_ylabel('PPF')


plt.tight_layout()
plt.show()
```
